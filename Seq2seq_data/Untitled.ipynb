```python
fr=open('chat_dnk.txt','r',encoding='utf-8')
Q_corpus=[]
A_corpus=[]
lines=fr.readlines()
for line in lines:
    Q_corpus.append(line[:line.find('\t')])
    A_corpus.append(line[line.find('\t')+1:])
fr.close()
```

```python
result=[]
num=0
for i in range(len(Q_corpus)):
    cha= A_corpus[i].count(' ') - Q_corpus[i].count(' ')
    if cha>0:
        print(i)
        num+=1
        a=cha
        idx=len(A_corpus[i])-1
        while a>=0:
            print(A_corpus[i][idx:])
            idx=A_corpus[i].rfind(' ',0,idx-1)
            a-=1
        
        result.append(Q_corpus[i][Q_corpus[i].rfind(' ')+1:]+'\t'+A_corpus[i][idx+1:])
    else:
        result.append(Q_corpus[i][Q_corpus[i].rfind(' ')+1:]+'\t'+A_corpus[i][A_corpus[i].rfind(' ')+1:])
        
```

```python
fw=open('chat_dnkend.txt','w',-1,'utf-8')
for line in result:
    fw.write(line)
fw.close()
```
